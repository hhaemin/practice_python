<br></br>
    - hint1: '표준 영문명'에서 찾지 못한다면 '구영문명'으로도 찾아보세요
<br></br>
    - hint2: 대문자-소문자 차이가 있을 수 있습니다.
<br></br>
    - 참고: 'Code'는 성분사전(Ingredients Dictionary)의 '성분코드'를 의미합니다.(Index가 아닙니다)
<br></br>
- 완료 후 결과 dataframe 변수를 check_03_01 함수에 입력하여 채점하세요.

````python
def get_code_list(df_target, df):
    ingredients_list = df_target['Ingredients List']
    ingredients = df['표준 영문명'] | df['구영문명']
    code_list = []
    for ingredient in ingredients_list:
        if ingredient in df.index:
            code_list.append(df.loc[ingredient]['성분코드'])
        else:
            ingredient_lower = ingredient.lower()
            if ingredient_lower in df.index:
                code_list.append(df.loc[ingredient_lower]['성분코드'])
    return code_list

df_target['Code List'] = df_target.apply(lambda df_target: get_code_list(df_target, df), axis=1)
df_target.head()
````
